```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename):
    data = np.loadtxt(filename, delimiter='\t')
    return data[:, 0], data[:, 1]
```

```python
i_almost_sorted, time_almost_sorted = load_data('almost_sorted_array_time.txt')
i_merge_insertion_almost_sorted, time_merge_insertion_almost_sorted = load_data('merge+insertion_almost_sorted_array_time.txt')
i_merge_insertion_random, time_merge_insertion_random = load_data('merge+insertion_random_array_time.txt')
i_merge_insertion_random_sorted, time_merge_insertion_random_sorted = load_data('merge+insertion_random_sorted_array_time.txt')
i_random, time_random = load_data('random_array_time.txt')
i_random_sorted, time_random_sorted = load_data('random_sorted_array_time.txt')
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(i_almost_sorted, time_almost_sorted, label='  ')
plt.plot(i_random, time_random, label=' ')
plt.plot(i_random_sorted, time_random_sorted, label='  ')
plt.title('Merge Sort')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(i_merge_insertion_almost_sorted, time_merge_insertion_almost_sorted, label='  ')
plt.plot(i_merge_insertion_random, time_merge_insertion_random, label=' ')
plt.plot(i_merge_insertion_random_sorted, time_merge_insertion_random_sorted, label='  ')
plt.title('Merge+Insertion Sort')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(i_almost_sorted, time_almost_sorted, label='Merge Sort')
plt.plot(i_merge_insertion_almost_sorted, time_merge_insertion_almost_sorted, label='Merge+Insertion Sort')
plt.title('  ')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(i_random, time_random, label='Merge Sort')
plt.plot(i_merge_insertion_random, time_merge_insertion_random, label='Merge+Insertion Sort')
plt.title(' ')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(i_random_sorted, time_random_sorted, label='Merge Sort')
plt.plot(i_merge_insertion_random_sorted, time_merge_insertion_random_sorted, label='Merge+Insertion Sort')
plt.title('  ')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()

plt.tight_layout()
plt.show()
```
